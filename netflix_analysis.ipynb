```python
#chcek missing values per column
df.isnull().sum()
```

```python
#For simplicity, fill missing values in some important columns
df['country']= df ['country'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Unknown')
df['duration'] = df['duration'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
```

```python
#drop rows where title is missing(very rare)
df = df.dropna(subset=['title'])
```

```python
df.head()
```

```python
#count of Movies vs TV Shows
type_counts = df['type'].value_counts()
print(type_counts)
```

```python
#percentage
(type_counts / len(df)) * 100
```

```python
#Some rows have multiple countries separated by commas
#we'll split and count

df_countries = df['country'].str.split(",", expand=True).stack()
df_countries = df_countries.str.strip()
country_counts = df_countries.value_counts().head(10)

country_counts
```

```python
#"Listed-in" column containd genres / categories
df_genres = df['listed_in'].str.split(',', expand=True).stack()
df_genres = df_genres.str.strip()
genre_counts = df_genres.value_counts().head(10)

genre_counts
```

```python
#convert date_added to string first
df['date_added'] = df['date_added'].astype(str)

# Clean & convert date_added to datetime
df['date_added'] = df['date_added'].str.strip()  
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extract year
df['year_added'] = df['date_added'].dt.year

# Count titles per year
titles_per_year = df['year_added'].value_counts().sort_index()
titles_per_year
```

```python
#All American movies
american_movies = df[(df['country'].str.contains("United States", na= False)) & (df['type'] == "Movie")]
american_movies.head()
```

```python
#All TV-MA rated shows
tv_ma = df[df['rating'] == "TV-MA"]
tv_ma.head()
```
